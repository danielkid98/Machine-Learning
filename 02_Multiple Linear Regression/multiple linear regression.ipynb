```python
import pandas as pd

data = pd.read_csv("Salary_Data2.csv")
data
```

可根據學歷高低來使用數字代表，但是對於城市民來講，因為不知道高低關係，所以無法用單一代號。

```python
data["EducationLevel"] = data["EducationLevel"].map({"高中以下": 0, "大學": 1, "碩士以上": 2})
data
```

```python
from sklearn.preprocessing import OneHotEncoder

onehot_encoder = OneHotEncoder()#因為此套件是一個類別，所以要先做轉換
onehot_encoder.fit(data[['City']])#傳入特徵，由於只接受傳入二維陣列，所以要改變形式
city_encoder = onehot_encoder.transform(data[['City']]).toarray()#開始轉換，並回傳完整的矩陣(因為預設為傳回稀疏矩陣) # type: ignore
city_encoder
```

```python
data[['CityA', 'CityB', 'CityC']] = city_encoder
data.drop(columns=['City', 'CityC'], inplace=True)
data
```

```python
from sklearn.model_selection import train_test_split

x = data.loc[:, ['YearsExperience', 'EducationLevel', 'CityA', 'CityB']]
y = data.loc[:, ['Salary']]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=87)#random_state可以固定分割的資料，確保不會一直變動

x_train = x_train.to_numpy()
x_test = x_test.to_numpy()
```

```python
import numpy as np

w = np.array([1, 2, 3, 4])
b = 1
y_predict = (x_train * w).sum(axis=1) + b
#如果axis等於1的話那就是進行每行的加總，如果等於0的話就是進行每列的加總
y_predict
```

```python
y_train = y_train.to_numpy()

((y_train - y_predict)**2).mean()#計算平均
```

```python
def compute_cost(x, y, w, b):
    '''
    returns:
    cost is corresponding w and b

    Parameters:
    x, y(training data)
    w, b(a array of weight, a constant of bias)
    '''
    y_predict = (x * w).sum(axis=1) + b
    cost = ((y - y_predict)**2).mean()

    return cost
```

```python
w = np.array([0, 2, 2, 4])
b = 0
compute_cost(x_train, y_train, w, b)
```

Gradient descent就是根據斜率改變參數(w1, w2, w3, w4, b)                           
1.可以透過微分的方式計算出一個點的切線斜率。                                      
如何計算出斜率?                                                                 
    先設定好cost function長什麼樣子，這邊的cost function就是(y - y_pred)**2      
    =(y - (w1*x1 + w2*x2 + w3*x3 + w4 * x4 + b))**2                              
    w1微分 = 2 * x1(y_predict - y)                                               
    W2微分 = 2 * x2(y_predict - y)                                               
    W3微分 = 2 * x3(y_predict - y)                                               
    W4微分 = 2 * x4(y_predict - y)                                               
    b微分 = 2(y_predict - y)                                                   
2.接著開始更新w以及b                                                            
    w1 - w1方向斜率 * 學習率                                                      
    w2 - w2方向斜率 * 學習率                                                      
    w3 - w3方向斜率 * 學習率                                                      
    w4 - w4方向斜率 * 學習率                                                      
    b - b方向斜率 * 學習率                                                        
3.接著一直重複更新w以及b知道找到最低的cost為止                                    

```python
y_predict = (x_train * w).sum(axis=1) + b
b_gradient = (y_predict - y_train).mean()

w_gradient = np.zeros(x_train.shape[1])
for i in range(x_train.shape[1]):
    w_gradient[i] = (x_train[:, i] * (y_predict - y_train)).mean()

w_gradient, b_gradient
```

```python
def compute_gradient(x, y, w, b):
    '''
    Returns:
    w_gradient(微分結果, a array), b_gradient(微分結果)
    
    Parameters:
    x, y(data)
    w, b(a array of weight, a constant of bias)
    '''
    y_predict = (x * w).sum(axis=1) + b
    b_gradient = (y_predict - y).mean()

    w_gradient = np.zeros(x.shape[1])
    for i in range(x.shape[1]):
        w_gradient[i] = (x[:, i] * (y_predict - y)).mean()

    return w_gradient, b_gradient
```

```python
w = np.array([1, 2, 2, 4])
b = 1
learning_rate = 0.001

w_gradient, b_gradient = compute_gradient(x_train, y_train, w, b)

print(compute_cost(x_train, y_train, w, b))

w = w - w_gradient * learning_rate
b = b - b_gradient * learning_rate
print(compute_cost(x_train, y_train, w, b))
print(w, b)
```

```python
np.set_printoptions(formatter={'float': '{: .2e}'.format})
#需要轉換格式因為w以及w_gradient是一個陣列形態
def gradient_descent(x, y, w_init, b_init, learning_rate, cost_function, gradient_function, run_iter, p_iter=1000):
    
    c_hist = []
    w_hist = []
    b_hist = []
    
    w = w_init
    b = b_init

    for i in range(run_iter + 1):
        w_gradient, b_gradient = gradient_function(x, y, w, b)

        w = w - w_gradient*learning_rate
        b = b - b_gradient*learning_rate
        cost = cost_function(x, y, w, b)

        w_hist.append(w)
        b_hist.append(b)
        c_hist.append(cost)

        if i%p_iter == 0:
            print(f"Iteration {i:5} : Cost {cost: .4e}, w: {w}, b: {b: .2e}, w_gradient: {w_gradient}, b_gradient: {b_gradient: .2e}")

    return w, b, w_hist, b_hist, c_hist
```

```python
w_init = np.array([1, 2, 2, 4])
b_init = 0
learning_rate = 1.0e-2
run_iter = 20000
p_iter = 2000

w_final, b_final, w_hist, b_hist, c_hist = gradient_descent(x_train, y_train, w_init, b_init, learning_rate, compute_cost, compute_gradient, run_iter, p_iter)
```
